```python
#Creado por Gabriel Missael Barco el 23/11/18

import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as con

%matplotlib inline 
```

```python
#Cargamos todos los puntos del archivo a un arreglo
puntos = np.loadtxt("IRCF.txt").T
#Guardamos el numero de puntos a tratar
N = np.size(puntos[1])
```

```python
#Esta función graficará todas las aproximaciones hechas con la ley de planck, ademas de los datos
#medidos experimental mente del fondo cosmico de microndas, con su incertidumbre
def grafica(b, a):
    #Creamos la figura, de tamaño deseado
    f = plt.figure(figsize=(12,12))
    ceros = np.zeros(N)
    
    #Creamos la primera subgrafica, la cual tendra el CMB y las aproximaciones
    ax1 = f.add_subplot(211)
    ax1.set_title('Fondo cosmico de microondas')
    ax1.scatter(puntos[0], puntos[1])
    for i in range (a):
        ax1.plot(puntos[0], b[i])
    ax1.grid()
    ax1.set_ylabel('Intensidad (MJy/sr)')
    ax1.set_xlabel('Frecuencia (1/cm)')
    
    #Esta segunda grafica tiene las incertidumbres del CMD
    ax2 = f.add_subplot(212)
    ax2.set_title('Incertidumbre de los datos medidos experimentalmente')
    ax2.errorbar(puntos[0], ceros, puntos[3]/1000)
    ax2.grid()
    ax2.set_xlabel('Frecuencia (1/cm)')
    ax2.set_ylabel('Incertidumbre (MJy/sr)')
    
    return
```

```python
#Esta funcion es la Distribución de Planck, la cual de acuerdo a una frecuencia y a una temperatura, 
#devuelve una intensidad. La usamos para conocer la temperatura de fondo del universo
def Int(v, T):
    h = con.h
    c = con.c
    k = con.k
    v *= (c*100)
    factor = (2*h*v**3/c**2)*1e26
    arg=(h*v)/(k*T)
    I = factor/(np.exp(arg)-1)
    return I/1e6
```

```python
#Esta funcion obtiene el promedio de minimos cuadrados para cada aproximación a cada temperatura
#Servira para conocer cual es la temperatura que mas se acerca
def X2(a):
    chi =0.0
    for i in range(N):
        chi += (puntos[1][i]-a[i])**2
    return chi
```

```python
#Esta funcion obtiene las N intensidades para una temperatura dada, usando la distribución de planck 
# de la funcion Int()
#Devuelve un arreglo
def temp(t):
    y=(np.zeros(N))
    for i in range (N):
        y[i] = Int((puntos[0][i]),t)
    return y
```

```python
#Creamos una matriz donde guardaremos cada una de las 11 aproximaciones que haremos 
res = np.zeros((11, N))
count =1
#Este ciclo obtendra 11 arreglos, de temperatura i*0.3, usando la funcion temp()
for i in range (11):
    res[i]=temp(0.3*count)
    count+=1
#Se grafican los 11 resultados obtenidos
grafica(res, 11)
```

```python
best = 1000000000.0
aux =0.0
pos = 0

#En este ciclo buscamos cual conjunto de puntos, de los 11 que obtuvimos, se aproxima
# mas a los datos del CMB
for i in range (11):
    aux=0
    aux = X2(res[i])
    if(aux<best):
        best = aux
        pos = i
        
#Imprimimos la temperatura que mas se aproxima
print ('Temperatura aproximada de fondo:', (pos+1.00)*0.300, '°K')
print('X²:', best)

#Graficamos ambas curvas, la conocida experimentalmente y la obtenica con la aproximacion
f2 = plt.figure(figsize=(12,8))
ax = f2.add_subplot(111)
ax.scatter(puntos[0], puntos[1])
ax.plot(puntos[0], puntos[1], label='CMB')
ax.plot(puntos[0], res[pos], 'r', label='2.7°K - Cuerpo negro')
ax.grid()
ax.set_ylabel('Intensidad (MJy/sr)')
ax.set_xlabel('Frecuencia (1/cm)')
plt.legend()
```
